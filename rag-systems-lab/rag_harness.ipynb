```python
import os
from dotenv import load_dotenv

load_dotenv()

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")

print("API Key loaded:","YES" if GEMINI_API_KEY else "NO check your env file")
```

```python
from llama_index.llms.gemini import Gemini
from llama_index.embeddings.gemini import GeminiEmbedding
from llama_index.core import Settings

llm = Gemini(
    api_key=GEMINI_API_KEY,
    model="gemini-2.5-flash"
)

embed_model = GeminiEmbedding(
    api_key=GEMINI_API_KEY,
    model_name="models/gemini-embedding-001"
)

Settings.llm = llm
Settings.embed_model = embed_model

print("LLM and Embeddings ready")
```

```python
from llama_index.core import VectorStoreIndex, Document

documents = [
    Document(text="The Eiffel Tower is located in Paris, France. It was built in 1889 by Gustave Eiffel."),
    Document(text="The Great Wall of China was built over many centuries, starting from the 7th century BC."),
    Document(text="Albert Einstein was born in Ulm, Germany in 1879. He developed the theory of relativity."),
    Document(text="The Amazon River is the largest river in the world by water discharge, located in South America."),
    Document(text="Python programming language was created by Guido van Rossum and released in 1991."),
]

index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

print("Index built successfully")
```

```python
import google.generativeai as genai

genai.configure(api_key=GEMINI_API_KEY)

for m in genai.list_models():
    if "embed" in m.name.lower():
        print(m.name)
```

```python
response = query_engine.query("Where is Effiel tower located?")
print("Response:", response)
```

```python
stress_test_queries = [
    # Factual - should answer correctly
    "Where is the Eiffel Tower located?",

    # Out of scope - should say it doesn't know
    "Who is the current president of France?",

    # Misleading - contains false premise
    "When was the Eiffel Tower built in London?",

    # Conflation - mixes two facts
    "Was the Amazon River discovered by Albert Einstein?",

    # Overreach - asking for info not in documents
    "What is the height of the Eiffel Tower?",

    # Negation trap
    "Was the Great Wall of China built after 1900?",
]

results = []

for query in stress_test_queries:
    response = query_engine.query(query)
    results.append({
        "query": query,
        "response": str(response)
    })

    print(f"Q: {query}")
    print(f"A: {response}")
    print("-" * 50)
```
