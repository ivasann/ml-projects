```python
missing = df.isna().sum()
missing_pct = (missing / len(df)) * 100
```

```python
missing_df = pd.DataFrame({'Missing_Values': missing,
                           "Percentage %": missing_pct.round(2)})
print(missing_df[missing_df['Missing_Values'] > 0])
```

```python
df['director'].fillna('Unknown', inplace= True)
df['cast'].fillna('Unknown', inplace= True)
df['country'].fillna('Unknown', inplace= True)
df['rating'].fillna('Unknown', inplace= True)

# Drop rows where 'date_added' or 'duration' is missing (very few)
df.dropna(subset= ['date_added', 'duration'], inplace= True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_year'] = df['date_added'].dt.month_name()
```

```python
print('Data Cleaned')
print(f'Remaning rows: , {df.shape[0]}')
```
